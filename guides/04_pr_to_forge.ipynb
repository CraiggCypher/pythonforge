## 🤝 Contributing to PythonForge

Thank you for contributing to PythonForge! Your work helps improve the course for every student.

---

## 📋 How Pull Requests Work

Since many students share this repository, we follow a safe contribution process:

1. **Fork** the repository — this creates your own copy at `https://github.com/your_username/pythonforge`
2. **Create a branch** for your changes:

```bash
git checkout -b your-branch-name
```

3. **Stage and commit** only the files you changed:

```bash
git add your_filename
git commit -m "brief description of your change"
```

4. **Push** to your fork:

```bash
git push origin your-branch-name
```

5. **Open a PR** on GitHub — go to the original PythonForge repo and click **"New Pull Request"**

---

## ✅ Before You Submit

Make sure your PR meets all of the following:

- [ ] Changes are **only inside** the `community/` contributions folder
- [ ] All Jupyter Notebook **outputs are cleared** before committing
- [ ] PR contains a **manageable number of files** (ideally under 10 files / 1,000 lines)
- [ ] No LLM-generated extras (verbose READMEs, unnecessary test files, `.env.example`, `requirements.txt`, etc.)
- [ ] No database files, personal files, or large data files

---

## 🔍 How to Check Your PR

After submitting, open your PR and click the **"Files Changed"** tab.

- Every changed file must be inside the `community/` folder
- The green number = lines added, the red number = lines deleted — **there should be no deleted lines**
- If any file shows **"Differences are too large to preview"**, your notebook outputs were not cleared

---

## ⚠️ Common Issues

**Issue 1 — Changes outside `community/`**

Only files inside the `community/` folder should appear in your PR. Changes elsewhere will block the merge.

**Issue 2 — Notebook outputs not cleared**

Before committing any `.ipynb` file, clear all outputs:

> In Jupyter: **Kernel → Restart & Clear Output**, then save

**Issue 3 — PR is too large**

If your project is 10+ files or 3,000+ lines, do this instead:

- Create a single `.md` or `.ipynb` file inside `community/`
- Describe your project and include a link to your GitHub repo where the full code lives

This keeps the repo lean and drives traffic to your profile.

**Issue 4 — LLM-generated extras**

Using AI to help build your project is encouraged — but you are the author. Remove:

- Overly long READMEs
- Masses of test files
- Heavily commented or overly defensive code
- Unnecessary config files

An ideal PR is a few focused files. Less is more.

**Issue 5 — Files that do not belong**

Do not include database files, source data, or personal files. Instead, add a note explaining how other students can generate those files themselves.
